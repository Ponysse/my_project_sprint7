```python
import pandas as pd
import plotly.express as px
import streamlit as st
```

```python
sleep_df = pd.read_csv('./student_sleep_patterns.csv')
print(sleep_df.sample(5))
```

```python
fig = px.histogram(sleep_df, x="Sleep_Duration", color_discrete_sequence=["slateblue"]) # crear un histograma

fig.update_layout(
    xaxis_title="Sleep duration",
    yaxis_title="Students",
    title="Sleep duration distribution"
)

fig.show()
```

```python
fig = px.scatter(
    sleep_df,
    x="Study_Hours",
    y="Sleep_Duration",
    color="Caffeine_Intake",
    color_continuous_scale="Viridis",
    title="Sleep vs. Study (color = caffeine)",
    labels={"Caffeine_Intake": "Caffeine intake"}
)

fig.update_layout(
    xaxis_title="Study Hours",
    yaxis_title="Sleep Duration",
)
fig.show()
```
